다음으로 데이터의 이상치를 확인해볼까요?  
(이상치: 잘못 평가된 값으로 잘못된 분석결과를 초래할 수 있는 값을 의미)  

**describe()**를 사용하면 데이터의 이상치를 알수 있습니다.

<font color='red'>- describe(): 각종 통계량을 요약해서 출력해주는 메서드</font>

```python
df_train.describe()
```

-> Count : 칼럼 별 총 데이터 수  
   mean / std : 칼럼 별 데이터의 평균 / 표준편차  
   min / max : 칼럼 별 데이터 최소값 / 최대값  
   25% / 50% / 75% : 백분위수의 각 지점으로, 분포를 반영해 평균을 보완하는 목적으로 사용
   
빅데이터 파악을 위해 많은 사람들이 기본 data.describe()을 사용하는 경우가 많습니다. 하지만 특정 칼럼 / 특정 시리즈의 요약정보를 확인하여 본다면 보다 자세히 알 수 있습니다.  
**df['칼럼명'].describe()**
   
- df['칼럼명'].describe(): 특정 컬럼 / 시리즈 요약

[Plus]
- df[['칼럼명','칼럼명']].describe()  
여러 개의 칼럼의 요약정보를 확인하고 싶으면 <span style="color:red">반드시 칼럼 전체를 리스트로 묶어 처리할 것</span>

위 내용의 이해를 돕기 위해 예시를 들어볼까요?

```python
df_train['Class'].describe()
```

-> Count: 총 데이터 수  
Unique: 중복없이 나오는 고유한 데이터 값 (Business, Eco plus, Eco임을 알 수 있었음)  
Top: 가장 값이 많은 데이터 (최빈값인 항목)  
Freq: 최빈 데이터의 실제 수 (Top의 개수, 최빈값)

[Plus]
- df['칼럼명'].unique(): Distinct 값 확인, 즉 중복 제외한 값의 종류 확인용(Unique와 동일)
- df['칼럼명'].value_counts(): 중복 포함해 전체 데이터 수 Count와 동일

# 데이터 시각화하기

```python
import missingno as msno
import matplotlib as mpl

msno.matrix(df_train,color=(0.1, 0.6, 0.8))
```

**Matrix**
msno.matrix(df_train,color=(0.1, 0.6, 0.8))  
df_train으로 지정한 트레인 데이터셋에 대한 결측치를 매트릭스로 시각화했습니다.  
color 파라미터는 RGB를 지정해주는 것입니다.

이 결과를 통해 df_train데이터는 24개의 column과 3000개의 row를 가지고 있다는 것을 확인했고, 결측치가 없다는 것을 다시 한번 확인했습니다.

만약 결측치가 있을경우 막대그래프에 흰색으로 표현되어집니다.
